```python
#I want to delete the first bird
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
im1=cv2.imread("/content/drive/MyDrive/Images for tests/birds")
cv2_imshow(im1)
```

```python
print(im1.shape)
```

```python
#first convert the image from BGR to GRAY to make mask
im1_gray=cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)

mask1=np.zeros(im1_gray.shape,dtype="uint8")
mask1[270:410,80:180]=255 

output=cv2.inpaint(im1,mask1,3,cv2.INPAINT_TELEA)
cv2_imshow(output)
```
